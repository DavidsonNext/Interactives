```python
%%HTML
<!DOCTYPE html>
<html>
    <head>
        <meta charset="UTF-8">
        <title>Money and the Money Market</title>
        <script type="text/javascript" src="https://cdn.mathjax.org/mathjax/latest/MathJax.js?config=AM_HTMLorMML-full"></script>
        <script type="text/javascript" src="https://cdnjs.cloudflare.com/ajax/libs/jsxgraph/0.98/jsxgraphcore.js"></script>
        <script type="text/javascript" src="underscore-min.js"></script>
        <script type="text/javascript" src="jquery.min.js"></script>
        <script type="text/javascript" src="jquery-ui.min.js"></script>
        <script type="text/javascript" src="../../JS/Macro.js"></script>
        <script type="text/javascript" src="MMM_Activity_3.js"></script>
    </head>

    <body>
        <!-- COMMENT: Define the jxgbox - aka, where all the interactive graphing will go. -->
        <div style="width: 100%; overflow: hidden;">
            <div id='jxgbox1' class='jxgbox' style='width:550px; height:450px; float:left; border: solid #1f628d 2px;'></div>
        </div>
        
        <div class="line">
            <input class="btn" type="button" value="Fall in Money Demand" onClick="increaseXY()">
            <input class="btn" type="button" value="Rise in Money Demand" onClick="decreaseXY()">
            <input class="btn" type="button" value="Reset" onClick="resetAnimation()">
        </div>
        
        
        <!-- COMMENT: Where our Javascript begins. -->
        <script type="text/javascript" src="https://cdnjs.cloudflare.com/ajax/libs/jsxgraph/0.98/jsxgraphcore.js"></script>
        <script type='text/javascript'>
        
            bboxlimits = [-1.6, 12, 12, -1.1];
            var brd1 = JXG.JSXGraph.initBoard('jxgbox1', {axis:false, 
                                                    showCopyright: false,
                                                    showNavigation: false,
                                                    zoom: false,
                                                    pan: false,
                                                    boundingbox:bboxlimits,
                                                    grid: false,
                                                    hasMouseUp: true, 
            });
            
            xaxis = brd1.create('axis', [[0, 0], [12, 0]], {withLabel: true, label: {offset: [320,-20]}});
            yaxis = brd1.create('axis', [[0, 0], [0, 12]], {withLabel: true, label: {offset: [-60,260]}});

            //Axes
            xaxis.removeAllTicks();
            yaxis.removeAllTicks();
            var ylabel = brd1.create('text',[-1.5,10,"Interest<br>Rate"],{fixed:true,fontsize:14,highlight:false});
            var xlabel = brd1.create('text',[8,-0.5,"Quantity of Money"],{fixed:true,fontsize:14,highlight:false});
            
            //Demand 1
            var D1 = createDemand(brd1,{name:'D1',color:'Gray'});
            D1.setAttribute({fixed:true, dash:1});
            var G = brd1.create('glider',[5.5,6.5,D1],{fixed:true,visible:false});
            
            ////////////
            // Fixed Dashed Lines for Board 1
            ////////////
            var dashD1 = createDashedLines2Axis(brd1,G,
                                              {fixed:true,
                                               withLabel:true,
                                               xlabel:'Y1',
                                               ylabel:'PL1',
                                               color:'Gray'});
            
            
            //Demand 2
            var D2 = createDemand(brd1,{name:'D2',color:'DodgerBlue'});
            D2.setAttribute({withLabel:false,offset:[125,-85]});

            //Glider along demand curve
            var G = brd1.create('glider',[5.5,6.5,D2],{name:'A',withLabel:false,fixed:true});
    
            ////////////
            // Draggable Dashed Lines for Board 1
            ////////////
            var dashD2 = createDashedLines2Axis(brd1,G,
                                              {fixed:false,
                                               withLabel:false,
                                               xlabel:'Y2',
                                               ylabel:'PL2',
                                               color:'DodgerBlue'});
            
            
            
                        
            toggleLabels = function(toggle) {
                dashD2.X1.setAttribute({withLabel:toggle});
                dashD2.Y1.setAttribute({withLabel:toggle});
                D2.setAttribute({withLabel:toggle});
            };
            
            //////////////////
            // Interactivity
            //////////////////
            brd1.on('move', function() {      
                //Moving Dashed Lines in Board 1
                dashD2.Y1.moveTo([0, G.Y()]);
                dashD2.Y2.moveTo([G.X(), G.Y()]);

                dashD2.X1.moveTo([G.X(), 0]);
                dashD2.X2.moveTo([G.X(), G.Y()]);
                brd1.update()
            });
            
            brd1.on('mousedown', function() {      
                toggleLabels(true);
                brd1.update()
            });
            
            //Animation for shifting curve SouthWest
            decreaseXY = function() {
                resetAnimation(10);
                var speed = 1000;
                toggleLabels(true);
                
                D2.point1.moveTo([D2.point1.X()-0.5,D2.point1.Y()-1.5],speed);
                D2.point2.moveTo([D2.point2.X()-0.5,D2.point2.Y()-1.5],speed);
                
                dashD2.Y1.moveTo([0, G.Y()-1.5],speed);
                dashD2.Y2.moveTo([G.X()-0.5, G.Y()-1.5],speed);

                dashD2.X1.moveTo([G.X()-0.5, 0],speed);
                dashD2.X2.moveTo([G.X()-0.5, G.Y()-1.5],speed);
                
                brd1.update();
            }
            
            //Animation for shifting curve NorthEast
            increaseXY = function() {
                var speed = 1000;
                resetAnimation(10);
                toggleLabels(true);
                
                D2.point1.moveTo([D2.point1.X()+0.5,D2.point1.Y()+1.5],speed);
                D2.point2.moveTo([D2.point2.X()+0.5,D2.point2.Y()+1.5],speed);
                
                dashD2.Y1.moveTo([0, G.Y()+1.5],speed);
                dashD2.Y2.moveTo([G.X()+0.5, G.Y()+1.5],speed);

                dashD2.X1.moveTo([G.X()+0.5, 0],speed);
                dashD2.X2.moveTo([G.X()+0.5, G.Y()+1.5],speed);
                
                brd1.update();
            }
            
            resetAnimation = function(speed) {
                toggleLabels(false);                
                D2.point1.moveTo([1,10],speed);
                D2.point2.moveTo([9,2],speed);
                
                dashD2.Y1.moveTo([0, 5.5],speed);
                dashD2.Y2.moveTo([5.5, 5.5],speed);

                dashD2.X1.moveTo([5.5, 0],speed);
                dashD2.X2.moveTo([5.5, 5.5],speed);
                
                brd1.update();
            }
            
            
        </script>
    </body>
</html>
```
